```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')


df = pd.read_csv('../data/allDataMean.csv', header=0)


df.columns = [
    'Current', 'Voltage', 'Power_Factor', 'Power', 'Frequency',
    'Active_Energy', 'ESP32_Temperature', 'CPU_Consumption',
    'CPU_Power_Consumption', 'CPU_Temperature', 'GPU_Consumption',
    'GPU_Power_Consumption', 'GPU_Temperature', 'RAM_Consumption',
    'RAM_Power_Consumption', 'RAM_Temperature'
]

df = df.apply(pd.to_numeric, errors='coerce')

print("Shape:", df.shape)
print("\nFirst 5 rows:")
df.head()
```

```python
df.info()
```

```python
df.describe().round(2)
```

```python
print("=== Missing Values ===")
print(df.isna().sum())

print("\n=== Zero Value Counts ===")
print((df == 0).sum())

print("\n=== Rows where Power = 0 ===")
print(len(df[df['Power'] == 0]))
```

```python
# Remove rows where Power = 0
df_clean = df[df['Power'] > 0].copy()

print("Original shape:", df.shape)
print("Cleaned shape:", df_clean.shape)
print("Rows removed:", len(df) - len(df_clean))
```

```python
plt.figure(figsize=(12, 4))

plt.subplot(1, 2, 1)
sns.histplot(df_clean['Power'], bins=50, color='steelblue')
plt.title('Distribution of Power (Watts)')
plt.xlabel('Power (Watts)')
plt.ylabel('Count')

plt.subplot(1, 2, 2)
sns.boxplot(y=df_clean['Power'], color='orange')
plt.title('Boxplot of Power (Watts)')
plt.ylabel('Power (Watts)')

plt.tight_layout()
plt.show()
```

```python
#correlation heatmap
plt.figure(figsize=(14, 10))
sns.heatmap(df_clean.corr(), 
            annot=True, 
            fmt=".2f", 
            cmap='coolwarm',
            center=0)
plt.title('Feature Correlation Matrix')
plt.tight_layout()
plt.show()
```

```python
top_features = ['Power_Factor', 'ESP32_Temperature', 
                'Voltage', 'CPU_Power_Consumption',
                'RAM_Power_Consumption']

fig, axes = plt.subplots(2, 3, figsize=(15, 8))
axes = axes.flatten()

for i, feature in enumerate(top_features):
    axes[i].scatter(df_clean[feature], df_clean['Power'], 
                   alpha=0.1, color='steelblue', s=1)
    axes[i].set_xlabel(feature)
    axes[i].set_ylabel('Power (Watts)')
    axes[i].set_title(f'{feature} vs Power')


axes[5].set_visible(False)

plt.tight_layout()
plt.show()
```

```python
zero_percent = (df_clean == 0).sum() / len(df_clean) * 100
zero_percent = zero_percent.sort_values(ascending=False)

plt.figure(figsize=(12, 5))
zero_percent.plot(kind='bar', color='tomato')
plt.title('Percentage of Zero Values Per Column')
plt.ylabel('Zero %')
plt.xlabel('Columns')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

print(zero_percent.round(2))
```

```python
plt.figure(figsize=(14, 4))
df_clean['Power'].iloc[:5000].plot(color='green')
plt.title('Power Consumption Over Time (First 5000 readings)')
plt.xlabel('Index (minutes)')
plt.ylabel('Power (Watts)')
plt.tight_layout()
plt.show()
```

```python
features_to_drop = [
    'Current',           # 0.04 correlation
    'Frequency',         # -0.01 correlation
    'Active_Energy',     # 0.00 correlation
    'GPU_Consumption',   # ~99% zeros
    'GPU_Power_Consumption',  # ~99% zeros + 0.15 correlation
    'GPU_Temperature'    # ~99% zeros
]

df_final = df_clean.drop(columns=features_to_drop)

print("Final dataset shape:", df_final.shape)
print("Remaining columns:", list(df_final.columns))

# Save clean dataset
df_final.to_csv('../data/clean_data.csv', index=False)
print("\n Clean dataset saved to data/clean_data.csv")
```
